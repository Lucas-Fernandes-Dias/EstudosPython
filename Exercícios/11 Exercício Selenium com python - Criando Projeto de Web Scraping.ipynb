### Web Scraping

```python
#importando bibliotecas úteis

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# atribuindo a váriavel link, o link a ser carregado

# RETIRAR O JOGO DA VELHA PRA RODAR link = 'https://pbdatatrader.com.br/jogosdodia'

#abrindo o site, através do webdriver

driver = webdriver.Chrome()
driver.get(link)

#verificando se o elemento foi carregado

while len(driver.find_elements(By.TAG_NAME, 'iframe')) == 0:
    time.sleep(1)

#entrando no 1ª iframe

iframe = driver.find_element(By.TAG_NAME, 'iframe')
driver.switch_to.frame(iframe)

#entrando no 2ª iframe

iframe_novo = driver.find_element(By.TAG_NAME, 'iframe')
driver.switch_to.frame(iframe_novo)

time.sleep(4)

#buscando a tabela no iframe

tabela = driver.find_element(By.CLASS_NAME, 'innerContainer' )
texto = tabela.text

# tratando o texto como string

lista_texto = texto.split('\n')

#pegando o dado desejado

lista_colunas = lista_texto[:11]
lista_values = lista_texto[11:]

#tratando a lista

dic = {}
values = []

for coluna in lista_colunas:
    dic[coluna] = []
    values = []       
    for v in range(0,275,11):
        values.append(lista_values[v])        
    dic[coluna] = values[:]
    lista_values = lista_values[1:]    

# criando o data frame no pandas

tabela_df = pd.DataFrame.from_dict(dic)
display(tabela_df)
    

```
